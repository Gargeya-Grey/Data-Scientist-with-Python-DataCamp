### For Loop on a DataFrame
We might think that, just like any other numpy array, if we use `for` loop on the dataframe to retrieve data, we can get it.<br>
`for val in df:
    print(val)`<br>
But this would print **only column names**.

For us to get the values of each row in a dataframe, we have to use a addtional function on the dataframe and then use a `for` loop on top of that. The method is: `df.iterrows()`. Using this we can get two values as a collection, 
- first one is the label
- second one is the series object containing the values in one row.

**SYNTAX**:
> `for lab,row in df.iterrows():
>    print(lab)
>    print(row) ## Series object`

Now, you can easily do *selective printing*, accessing the elements of the series object of the row.(like we discussed above, column access)

---
#### Adding a Column in a dataframe:
Here, we will be discussing two specific ways for doing that and their pros and cons.<br>**FIRST**

> `for lab, row in df.iterrows():
        df.loc[lab, "New_Column_name"] = len(row['Country'])`<br>
    Using this method we are putting the length of string values of the country column for each row in a new column.
    **Note**: Here we are accessing AND updating the values in the dataframe at each iteration. *now* for a smaller dataset,this is no big deal. But for larger dataset. This is very **inefficient**.

**SECOND**
> `df['New_column_name'] = df['Country'].apply(len)`
<br> Now such a way is way simpler to read, much faster and much **more efficient**.
By using the apply function on the pandas series, retieved from column access method, and then applying a function on all its value altogther is much computationally faster. Then we simply created a new column name and update its value with the resultant series we got earlier.

## Unit 5: Case Study :: Hacker Statistics
---

### Question: 
You throw a dice, while walking up to the top of empire state building. You throw that dice a 100 times. If the dice turns 1 or 2 you step down 1 step, if your dice turns 3 or 4 or 5 you step up 1 step. But if its 6 on the dice, you throw the dice again and you walk up the number the dice shown after that roll. 
- You can't go below step 0
- 0.1 % chance of falling down the stairs to the very start
- **BET:** you'll react 60 steps higher, in those 100 dice throws
---

### How to Solve?
* Analytical calculations for the accuracte and dead precise probability.
* Simulate the process so many times that we get a clear hang of fractions that we react 60 steps in 100 dice throws from those collective simulations

**The Second Approach is called _Hacker Statistics!_**

---
The Things we need in order to set this whole up are:
1. **Random Generators**: We need random numbers in our simulations to get the unbiased and averaged out results in our probabilities in the end. We do this with the help of _Numpy Module's Random API_. This is technically a psuedo-random generator where psuedo stands for the computer algorithms that generates the random number at the very base, so technically it's not completely random and we can set the **seed** to generate the same sequence of results everytime with the same seed. This ensures **Reproducibility**. `np.random.rand()` for getting the randome number between 0 to 1, `np.random.seed(123)` for setting the seed, for consistent between runs.
2. **Random Walk**: When you throw a dice, you get a random result so called a _random step_. But we repeat this same process successively, the previous values are addded to the new ones/ the new values are calculated on top of the old ones. This is what we call a **Random Walk**. To record every step in your Random walk, we need to know to gradually build a list. This concept is well known in science, For eg:
    - A path travelled by a molecule
    - Gambler's financial status


```python
import numpy as np

np.random.seed(123)

# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice >2 and dice <6 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice)
print(step)
```

```python
import numpy as np

np.random.seed(123)

# Initialize random_walk
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        # Replace below: use max to make sure step can't go below 0
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

print(random_walk)

### ADDING THE VISUALIZATION

import matplotlib.pyplot as plt

plt.plot(random_walk)
plt.show()
```
