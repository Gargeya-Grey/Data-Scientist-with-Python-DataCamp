```python
df = df.reset_index()
```

```python
# FacetGrid Object
import matplotlib.pyplot as plt
## Height in Inches for each Facet and aspect*height == width of each facet in inches.
g = sns.catplot(x="number", y="mass", data=df, kind="bar", height=8, aspect=1.5)
g.fig.suptitle(" for each planet")
plt.tight_layout()
plt.savefig("7.jpg", dpi=100)
plt.show()

```
